<b>1. </b>Load image from a file and display the image. Determine the size of the image. Finally, save a new copy of the image in a new file.

<b>2. </b>Write a function <code>flip_image</code>, which flips an image either vertically or horizontally. The function should take two input parameters: the matrix storing the image data and a flag to indicate whether the image should be flipped vertically or horizontally. Use this function to flip an example image both vertically and horizontally. <b>Tip:</b> You can use numpy array indexing or OpenCV's <a href="http://docs.opencv.org/modules/core/doc/operations_on_arrays.html#flip">flip</a> function to solve this problem.

```python
def flip_image(image,flag):
    """ Function that flips an image.
    Input:  image: numpy array with the image data.
    ======  flag: boolean, if True->Horizontal, if False->Vertical.
    
    Output: numpy array with the flipped image.
    ======
    """
    return np.flip(image,flag) # Using opencv: cv2.flip(image,flag)

for i, img in enumerate(images):
    horizontal = flip_image(img, True); # Call the function to flip 'img' horizontally
    vertical = flip_image(img, False); # Call the function to flip 'img' vertically
    fig = plt.figure();
    plt.subplot(1,3,1); plt.imshow(img, cmap='gray'); plt.title('Original Image'); # Subplot original image
    plt.subplot(1,3,2); plt.imshow(horizontal, cmap='gray'); plt.title('Horizontal Image'); # Subplot horizontal image
    plt.subplot(1,3,3); plt.imshow(vertical, cmap='gray'); plt.title('Vertical Image'); # Subplot vertical image
    fig.tight_layout()
    plt.show()
```

<p><strong><em>Code functionality:</em></strong> An image is flipped horizontally and then vertically. The original and both of the flipped results are displayed. A loop contains this so that several images go through this process.</p>
<p><strong><em>Functions used and description: </em></strong></p>
<p>The function <strong>np.flip()</strong>, flips the image horizontally or vertically depending on the flag that is used (which is the second argument it requires). A True or a 1 will cause a horizontal flip, whereas a false or a 0 will cause a vertical flip. The first argument that must be provided is the image itself.</p>
<p>Another possible option: instead of needing the flip function that requires a flag to know the type of flip that is going to be made, the function <strong>flipud</strong> (flip up-down) flips arrays vertically and the function <strong>fliplr</strong> (flip left-right) flips them horizontally.</p>


<b>3. </b> Write a function to generate the negative of an image. This means that  a new image is created in which the pixel values are all equal to 1.0 minus the pixel value in the original image.

```python
def negative(image):
    """ Function that generates the negative of an image.
    Input:  image: numpy array with the image data.
    =======
    
    Output: numpy array with the negative image.
    =======
    """
    return (255-image)

for img in images:
    neg_image = negative(img) # Call the function to get the negative image of 'img'
    fig = plt.figure();
    plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Original Image'); # Subplot the original image
    plt.subplot(1,2,2); plt.imshow(neg_image, cmap='gray'); plt.title('Negative Image'); # Subplot the negative image
    fig.tight_layout()
    plt.show()
```

<p><strong><em>Code functionality:</em></strong> The negative of an image is obtained thanks to mathematic operations. Both the original and the negative images are displayed. A loop contains the mention so that several images are processed.</p>
<p><strong><em>Functions used and description: </em></strong></p>
<p>According to the guidelines, the negative of an image can be obtained by operating 1 minus the pixel value in the original image. As our matrices are not normalized the values of the pixels go from 0 to 255, so in order to get the &ldquo;negative&rdquo; of the image a subtraction of 255-image is done. That is saved as neg_image. Both the negative and the original picture are displayed.</p>

<b>4. </b>Write a function <code>average_intensity</code>, which calculates the average intensity level of an image. Use this function on example images and discuss your results. You can use images from section 2 and 3.

```python
def average_intensity(image):
    """ Function that calculates the average intensity level of an image.
    Input:  image: numpy array with the image data.
    =======
    
    Output: numpy array with the average of the image.
    =======
    """
    return np.mean(image)

for img in images:
    print "Average intensity level of the image: " + str(average_intensity(img)) # Call and print the output of the function
```

<p><strong><em>Code functionality:</em></strong>&nbsp; The mean or average, of the values of the pixels of an image is calculated and written down. A loop is done in order for all the images to go through the code.</p>
<p><strong><em>Functions used and description: </em></strong></p>
<p>To get the average intensity level of an image the function <strong>np.mean()</strong> was used because it computes an arithmetic mean.</p>
<p>There is a similar function called <strong>np.average()</strong>, but the latter can compute a weighted average if provided the weights parameter, and also it doesn&rsquo;t take into account masks so the whole set of data is used to compute the average. On the other hand, np.mean() takes into account masks, so only unmasked values are used. Having this information, it can be concluded that for the purpose of this exercise using either np.average or np.mean would have had the same results.</p>
<p>As mentioned before, the values of the pixels are not normalized which mean they variate between 0 and 255. As expected, the images with paler shades of gray and white have the highest numbers in the average intensity level.</p>

<b>5. </b>Write a function <code>threshold_image</code> which thresholds an image based on a threshold level given as a parameter to the function. The function should take two parameters: the image to be thresholded and the threshold level. The result of the function should be a new thresholded image. <br />

```python
def threshold_image(image, th):
    """ Function that thresholds an image with an input threshold level.
    Input:   image: numpy array with the image data.
    =======     th: thershold value
    
    Output: img_th: numpy array with the thresholded image.
    =======
    """
    _, img_th = cv2.threshold(image, th, 255, 0)
    return img_th

fig = plt.figure();
for i, img in enumerate(images):
    th_image = threshold_image(img, 100) # Call the function to threshold 'img' with a th level of 100
    plt.subplot(3,3,1+i); plt.imshow(th_image, cmap='gray'); # Subplot the thresholded image
fig.tight_layout()
plt.show()
```

<p><strong><em>Code functionality:</em></strong>&nbsp; An image is turned to a black and white version of it. A threshold value is chosen so that all the values above it are the same, and so the values below it, get the opposite measure from the ones above. This binary image is then displayed. This code is inside of a loop that allows several images to have this modification.</p>
<p><strong><em>Functions used and description: </em></strong></p>
<p>The function <strong>cv2.threshold</strong> needs the source image as its first argument, it must be in a grayscale. The second argument is the threshold value used to classify the pixel values. A third argument defines the value to be given if the pixel value is higher than the threshold value. The fourth and final parameter indicates which of the 5 different styles of threshold is going to be used. For this exercise it was used the first (cv2.THRESH_BINARY, which is binary) one because the function in general is mostly used to get the binary image of the input image <a href="http://marker.to/fQ0tk9">(Reference 1)</a>. The second type of threshold is the inverted binary (cv2.THRESH_BINARY_INV). The third is the truncated one (cv2.THRESH_TRUNC), which leaves the pixel intensities of the image the same if they don&rsquo;t surpass the threshold value. The fourth (cv2.THRESH_TOZERO), gets the pixel to zero if it is not higher than the threshold value. The fifth (cv2.THRESH_TOZERO_INV), is the inverse version of the fourth type of threshold.</p>
<p>In the example the threshold value chosen was of 100.</p>

<b>7. </b>Write a function which subsamples a grayscale image of size 512x512 by factors of 2, i.e., 256, 128, 64 and display your results. There are multiple ways to do this as discussed in the textbook. You may simply sub-sample, average, etc.  Describe which you used and why.

```python
def subsample(image):
    """ Function that subsamples an image by factors of 2 until its factor=2^7=128.
    Input:  image: numpy array with the image data.
    =======
    
    Output:     a: list of numpy arrays with the resized images
    =======
    """
    a = []
    for i in range (8):
        factor = 2**i
        a.append(cv2.resize(image,None,fx=1./factor,fy=1./factor,interpolation=cv2.INTER_AREA))
    return a

for k, img in enumerate(images):
    print filenames[k]
    fig = plt.figure();
    for k2,i in enumerate(subsample(img)): # Call the function and get each image 'i' from the output list
        plt.subplot(3,3,1+k2); plt.imshow(i, cmap='gray'); plt.title(i.shape); # Subplot each resized image
    fig.tight_layout()
    plt.show()
```

<p><strong><em>Code functionality: </em></strong>There is a loop that subsamples an image making it look more pixelated every time. This process is done 8 times per image with a growing subsampling. Each of the results are added to an image that is going to display these 8 images. Then a loop that contains the latter allows another image to enter and receive the same processing. Creating a picture of 8 subsampled varieties for each image and displaying them.</p>
<p><strong><em>Functions used and description: </em></strong></p>
<p>The function <strong>cv2.resize() </strong>changes the scale of an image. The size can be specified manually or the scaling factor. It has as its first argument the input image. It second argument is the output image size. In the third and fourth argument, the numbers indicate the scaling factor for each axis. Lastly it is indicated the interpolation method, which for this case was CV2.INTER_AREA, because the internet documentation indicates that this is the one with the best results when shriking an image <a href="http://marker.to/EosMxQ">(Reference 2)</a> <a href="http://marker.to/xD0uBL">(Reference 3)</a>. <b>8. </b>Keeping your original image resolution reduce the number of gray levels in your image from 256 to 2 in powers of 2. Display your results.

```python
def rescale(image, factor):
    """ Function that reduces the number of gray levels in an image by factors of powers of 2.
    Input:    image: numpy array with the image data.
    =======  factor: integer number
    
    Output: img_out: numpy array of the image with the gray scale reduced
    =======
    """
    img = image/np.power(2,factor) # Reduce the pixel levels to a max value of 2^factor
    img_out = (img.astype('float')/np.max(img))*255 # Rescale the pixel values to an [0,255] scale
    return img_out
        
for img in images:
    fig = plt.figure();
    for i in range(1,8):
        img2 = rescale(img,i) # Call the function to get the image with 2^(8-i) gray levels
        plt.subplot(3,3,i); plt.imshow(img2, cmap='gray'); plt.title("#Lv="+str(2**(8-i)));
    fig.tight_layout()
    plt.show()
```

<p><strong><em>Code functionality:</em></strong>&nbsp; There is a loop where the exponent of the number 2 increases. In it, an image is divided by this power. The image is rescalated so that its values are distributed between 0 and 255 and not only in the limited range caused by the division. This is done normalizing the image previously scaled, by multiplying it by the maximum number of the new range, which is 255. Then all the images obtained in the loop are displayed. This loop is contained into another that changes the image so that several images go through this modification.</p>
<p><strong><em>Functions used and description: </em></strong></p>
<p>The function <strong>np.power() </strong>elevates the first argument to an exponent which is the second argument it requires. In this case the number 2 is going to be elevated by a factor that is going to increase due to a loop.</p>
<p>The function img.<strong>astype(&lsquo;float&rsquo;) </strong>&nbsp;allows the result of the division to be a decimal number. It is necessary because python2 gives integer numbers by default.</p>
<p>The function <strong>np.max()</strong> gives the maximum value in the image.</p>
